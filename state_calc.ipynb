### expected value calculation

```python
import numpy as np
def expected_value (values, weights):
    values = np.array(values)
    weights = np.array(weights)
    return ((values*weights).sum()/weights.sum())


Values = [1,8,9,7,6,2]
Probability = [0.1,0.6,0.4,0.9,0.7,0.3]   

print(expected_value(Values,Probability))
```

### Binomial Distribution Example

And the formula for the binomial cumulative probability function is: for n times trial for getting k times and each times getting probability is p

F(k;n,p)=∑(k=0 to x) [{n!/(n−k)!*k!}*{p^k*(1−p)^(n−k)}]


```python
# for a dice rolling for 12 times and getting a certain value (1-6) for 5 times where probability of each value is 1/6=0.17
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
n = 12
p = 0.17
x = np.arange(0, n+1) 
binomial_pmf = binom.pmf(x, n, p)     #pmf means probability mass function

print(binomial_pmf)

plt.plot(x, binomial_pmf, color='blue')
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

The graph shows that if we choose any number from 1 to 6 (die sides) and roll the die 12 times, the highest probability for any of those numbers to be observed is 2 times.

In other words, if I choose number 1 and roll the die 12 times, most likely 1 will show up 2 times.
