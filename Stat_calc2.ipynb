### Poisson's Distribution
The PMF (probability mass function) of a Poisson distribution is given by:

p(k,λ)={(λ^k)*(e^−λ)/k!}  

where:

    λ is a real positive number given by λ=E(X)=μ
    k is the number of occurrences
    e=2.71828

```python
# on avrg λ=7 hurrican happend on a country, then what is the probability of happening k=5 hurrican this year
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
k = np.arange(0, 17)
λ=7
pmf = poisson.pmf(k, λ)
pmf = np.round(pmf, 5)
for val, prob in zip(k,pmf):
    print(f"k-value {val} has probability = {prob}")
plt.plot(k, pmf, marker='o')
plt.xlabel('k')
plt.ylabel('Probability')
plt.show()
cdf = poisson.cdf(k, mu=7)       #Poisson CDF (cumulative distribution function) in Python
cdf = np.round(cdf, 3)
for val, prob in zip(k,cdf):
    print(f"k-value {val} has probability = {prob}")
plt.plot(k, cdf, marker='o')
plt.xlabel('k')
plt.ylabel('Cumulative Probability')

plt.show()



```

what if we wanted to find out the probability of seeing up to 5 hurricanes (mathematically: k≤5), we can see that it’s 0.30071 or 30.07%.

On the other hand, we can be interested in probability of observing more than 5 hurricanes (mathematically: k>5), which would be 1−p(5,7)=1−0.30071=0.69929 or 69.93%.
