### Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Create a normal distribution with mean as 5 and standard deviation as 10
mu = 5
std = 20
snd = norm(mu, std)
# Generate 1000 random values between -100, 100
x = np.linspace(-100, 100, 1000)
#
# Plot the standard normal distribution for different values of random variable
# falling in the range -100, 100
plt.figure(figsize=(7.5,7.5))
plt.plot(x, snd.pdf(x))             #pdf=probability distrubuted function
plt.xlim(-60, 60)
plt.title(f'Normal Distribution (Mean = {mu}, STD = {std})', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

here mean is 5, so the peak is slightly right side of 0 in the fig, again std is 20 which means 68% of data fall between +20 to -20 from the mean (sigma-1), 95% data fall between +40 to -40 (sigma-2) and 99.7% data fall between +60 to -60 (sigma-3).
